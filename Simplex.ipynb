```python
c_vec = input("Objective Function Coefficients, in Array Format (OBJ) : ")
a_matrix = input("Constraint Coefficients, in Matrix Format (LHS) : ")
b_vec = input("Constraints Values, in Array Format (RHS) : ")

obj = eval(c_vec)
lhs = eval(a_matrix)
rhs = eval(b_vec)
```

```python
# Using SciPy Library
from scipy.optimize import linprog
import numpy as np

optimize = linprog(c = [-i for i in obj],
                  A_ub = lhs,
                  b_ub = rhs,
                  method = 'simplex')
print(optimize)
```

```python
# Using LipPy Library
import lippy as lp

simplex = lp.SimplexMethod(obj, lhs, rhs)
solution, func_value = simplex.solve()

print("Solution : ", solution)
print("Objective Function : ", func_value)
```
