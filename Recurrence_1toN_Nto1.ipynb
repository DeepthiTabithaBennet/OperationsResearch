```python
def print1toN(n):
    if n > 0:
        print1toN(n - 1)
        print(n, end=' ')
```

```python
def printNto1(n):
    if n > 0:
        print(n, end=' ')
        printNto1(n - 1)
```

```python
n = int(input("Enter the value of N : "))
```

```python
print1toN(n)
```

```python
printNto1(n)
```
