```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
print("Enter the coefficients of the Polynomial, in the format ax + b")
a = int(input("Enter the Value of a : "))
b = int(input("Enter the Value of b : "))

num_data = int(input("\nEnter the Number of Data Points : "))
```

```python
# Generate sample data
np.random.seed(0)

# if x and y is not given in the question
x = np.linspace(-5, 5, num=num_data)
y = (a*x) + b + np.random.normal(scale=3.0, size=num_data)     # y = 2.0 + 1.5 * x + 3.0 * x**2 + np.random.normal(scale=3.0, size=x.shape)

# if a table of x and y is given in the question
x = np.array([1, 2, 3, 4 ,5])
y = np.array([3, 4, 5, 8, 10])
```

```python
# Define the nonlinear function
def func(x, a, b):
    return (a*x) + b
```

```python
# Fit the nonlinear model
popt, pcov = curve_fit(func, x, y)
```

```python
# Visualize the results
plt.scatter(x, y, label='Data')
plt.plot(x, func(x, *popt), 'red', label='Fit')
plt.legend()
plt.show()
```
