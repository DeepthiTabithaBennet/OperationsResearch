```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# if x and y is not given in the question
np.random.seed(0)
num_data = int(input("\nEnter the Number of Data Points : "))

x = np.linspace(-5, 5, num=num_data)
y = np.random.randint(-5, 5, size=num_data)

#--------------------------------------------------------------

# if a table of x and y is given in the question
x = np.array([1, 2, 3, 4 ,5])
y = np.array([3, 4, 5, 8, 10])
```

```python
# Define the nonlinear function
def func(x, a, b):
    return (a*x) + b
```

```python
# Fit the nonlinear model
popt, pcov = curve_fit(func, x, y)
```

```python
# Visualize the results
plt.scatter(x, y, label='Data')
plt.plot(x, func(x, *popt), 'red', label='Fit')
plt.legend()
plt.show()
```
