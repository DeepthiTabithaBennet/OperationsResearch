```python
import transportation_problem as tp

num_s = int(input("\nEnter the number of Supply Elements : "))
s = []

for i in range(num_s):
    key = input("Enter key: ")
    value = int(input("Enter value: "))
    s.append((key, value))


num_d = int(input("\nEnter the number of Demand Elements : "))
d = []

for i in range(num_d):
    key = input("Enter key: ")
    value = int(input("Enter value: "))
    d.append((key, value))


print("\nEnter the Cell Values : ")
c = []
intermed = []
for i in range(0, num_s):
    print("Enter Row", i, ": ")
    for j in range(0, num_d):
        x = int(input())
        intermed.append(x)
    c.append(intermed)
    intermed = []
```

```python
# Without MODI Method
p  =  tp.TransportationProblem(s, d, c)
r  =  p.solve(tp.VogelIniter)
print (r)
```

```python
# With MODI Method
p  =  tp.TransportationProblem(s, d, c)
r  =  p.solve(tp.VogelIniter, tp.PotentialChecker, tp.ClosedLoopAdjustmentOptimizer)
print (r)
```
