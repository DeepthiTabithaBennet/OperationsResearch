```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
```

```python
print("Enter the coefficients of the Polynomial, in the format ax1 + bx1 + c")
a = int(input("Enter the Value of a : "))
b = int(input("Enter the Value of b : "))
c = int(input("Enter the Value of c : "))

num_data = int(input("\nEnter the Number of Data Points : "))
```

```python
# Generate sample data
np.random.seed(0)

x1 = np.linspace(-5, 5, num=num_data)
x2 = np.random.normal(loc=0, scale=1, size=num_data)  # Additional independent variable

y = (a*x1) + (b*x2) + c + np.random.normal(scale=3.0, size=num_data)
```

```python
# Define the nonlinear function
def func(X, a, b, c):
    return (a*X[:,0]) + (b*X[:,1]) + c
```

```python
# Fit the nonlinear model
X = np.vstack((x1, x2)).T  # Stacking x1 and x2 as independent variables
popt, pcov = curve_fit(func, X, y)

print(popt)
print(pcov)
```

```python
# Visualize the results
ax = plt.axes(projection='3d')

ax.scatter3D(x1, x2, y, label='Data');
ax.plot_trisurf(x1, x2, func(X, *popt), alpha = 0.25, color='red', label='Fit')

plt.xlabel('x1')
plt.ylabel('x2')

plt.title('Multiple Linear Regression')
# plt.legend()
plt.show()
```
