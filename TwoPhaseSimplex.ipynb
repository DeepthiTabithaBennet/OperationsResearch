```python
from scipy.optimize import linprog
import numpy as np
```

```python
obj_num = int(input("Enter the number of variables in the Objective Function : "))
obj = []

print("Enter the Coefficients of the variables in the Objective Function : ")
for i in range(0, obj_num):
    x = int(input())
    obj.append(x)

print("\nObjective Function : ", obj)

# ---------------------------------------------------------------------------------------

const_num = int(input("\n\nEnter the number of Constraint Equations : "))
lhs = []
rhs = []
intermed = []

print("Enter the LHS of the Constraint Equations : ")
for i in range(0, const_num):
    print("\nConstraint Equation", (i+1))
    for j in range(0, obj_num):
        y = int(input())
        intermed.append(y)
    lhs.append(intermed)
    intermed = []

print("\nConstraint Equations (LHS) : ", lhs)

# ---------------------------------------------------------------------------------------

print("\n\nEnter the RHS of the Constraint Equations : ")
for i in range(0, const_num):
    x = int(input())
    rhs.append(x)

print("\nConstraint Equations (RHS) : ", rhs)

# ---------------------------------------------------------------------------------------

print("\n\n")
```

```python
optimize = linprog(c = obj,
                  A_ub = lhs,
                  b_ub = rhs,
                  # bounds = (0, float('inf')),
                  method = 'revised simplex')
print(optimize)
```
