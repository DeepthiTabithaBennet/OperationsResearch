```python
from munkres import Munkres, print_matrix
```

```python
row_num = int(input("Enter the number of Rows : "))
col_num = int(input("Enter the number of Columns : "))

matrix = []
intermed = []

print("\nEnter the Matrix : ")
for i in range(0, row_num):
    print("\nRow", (i+1))
    for j in range(0, col_num):
        y = int(input())
        intermed.append(y)
    matrix.append(intermed)
    intermed = []
```

```python
m = Munkres()
indexes = m.compute(matrix)

print_matrix(matrix, msg='Input Matrix :')
print_matrix(m.pad_matrix(matrix), msg='\nLowest cost through this matrix :')

print("\nOptimal Assignment : ")

total = 0
for row, column in indexes:
    value = matrix[row][column]
    total += value
    print('(', row, ',', column, ') ->', value)

print('\nTotal Cost :', total)
```
