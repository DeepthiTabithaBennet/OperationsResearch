```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
print("Enter the coefficients of the Polynomial, in the format ax^2 + bx + c")
a = int(input("Enter the Value of a : "))
b = int(input("Enter the Value of b : "))
c = int(input("Enter the Value of c : "))

num_data = int(input("\nEnter the Number of Data Points : "))
```

```python
# Generate sample data
np.random.seed(0)

x = np.linspace(-5, 5, num=num_data)
y = (a*(x**2)) + (b*x) + c + np.random.normal(scale=3.0, size=num_data)
```

```python
# Define the nonlinear function
def func(x, a, b, c):
    return (a*(x**2)) + (b*x) + c
```

```python
# Fit the nonlinear model
popt, pcov = curve_fit(func, x, y)
```

```python
# Visualize the results
plt.scatter(x, y, label='Data')
plt.plot(x, func(x, *popt), 'red', label='Fit')
plt.legend()
plt.show()
```
