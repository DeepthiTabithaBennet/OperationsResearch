```python
def conditional(p, q):
    """Evaluates truth of conditional for boolean variables p and q."""
    return (0 if p and not q
            else 1)


def biconditional(p, q):
    """Evaluates truth of biconditional for boolean variables p and q."""
    return (1 if p and q
            else 1 if not p and not q
            else 0)
```

### **Prove [(A→B)∧A]→B is a Tautology**

```python
print("|  A  |  B  |  A → B  |  (A → B) ∧ A  |  S  |", )

for i in range(0, 4):
    a = i // 2
    b = i % 2
    print("| ", a, " | ", b, " |   ", conditional(a, b), "   |      ", (conditional(a, b) and a), "      | ", (conditional((conditional(a, b) and a), b)), " |")
```

### **Prove (AVB)∧[(¬A)∧(¬B)] is a Contradiction**

```python
print("|  A  |  B  |  A V B  |  ¬A  |  ¬B  |  (¬A) ∧ (¬B)  |  S  |", )

for i in range(0, 4):
    a = i // 2
    b = i % 2
    print("| ", a, " | ", b, " |   ", (a or b), "   |  ", int(not a), " |  ", int(not b), " |      ", int((not a) and (not b)), "      | ", (a or b) and (int((not a) and (not b))), " |")
```

### **Prove (AVB)∧(¬A) is a Contingency**

```python
print("|  A  |  B  |  A V B  |  ¬A  |  S  |", )

for i in range(0, 4):
    a = i // 2
    b = i % 2
    print("| ", a, " | ", b, " |   ", (a or b), "   |  ", int(not a), " | ", ((a or b) and int(not a)), " |")
```
