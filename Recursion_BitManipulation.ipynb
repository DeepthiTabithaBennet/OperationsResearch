```python
# Function to return a^n
def powerOptimised(a, n):

    # Stores final answer
    ans = 1

    while (n > 0):
        last_bit = (n & 1)

        # Check if current LSB is set
        if (last_bit):
            ans = ans * a

        a = a * a

        # Right shift
        n = n >> 1

    return ans
```

```python
a = 3
n = 5

print(powerOptimised(a,n))
```
