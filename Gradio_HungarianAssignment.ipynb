```python
from munkres import Munkres, print_matrix
import gradio as gr
```

```python
def result(matrix):

    matrix = eval(matrix)

    m = Munkres()
    matrix_cost = munkres.make_cost_matrix(matrix)
    indexes = m.compute(matrix_cost)

    output = "Optimal Assignment : "

    total = 0
    for row, column in indexes:
        value = matrix[row][column]
        total += value
        output += ('\n(' + str(row) + ',' + str(column) + ') ->' + str(value))

    output += ('\n\nTotal Cost :' + str(total))

    return output
```

```python
iface = gr.Interface(
    fn=result,
    inputs=[
        gr.Textbox(label="Task Matrix"),
    ],
    allow_flagging="never",
    outputs="text",
    title="Hungarian Assignment Problem Solver"
)

iface.launch()
```
