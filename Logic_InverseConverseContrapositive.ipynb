```python
def conditional(p, q):
    """Evaluates truth of conditional for boolean variables p and q."""
    return (0 if p and not q
            else 1)


def biconditional(p, q):
    """Evaluates truth of biconditional for boolean variables p and q."""
    return (1 if p and q
            else 1 if not p and not q
            else 0)
```

### **Statement : “If you do your homework, you will not be punished.”**

A : Doing homework

B : Getting punished


### **Statement : [A → (¬B)]**

```python
print("|  A  |  B  |  ¬B  |  S  |", )

for i in range(0, 4):
    a = i // 2
    b = i % 2
    print("| ", a, " | ", b, " |  ", int(not b), " | ", (conditional(a, int(not b))), " |")
```

### **Inverse : [(¬A) → B]**

```python
print("|  A  |  B  |  ¬A  |  S  |", )

for i in range(0, 4):
    a = i // 2
    b = i % 2
    print("| ", a, " | ", b, " |  ", int(not a), " | ", (conditional(int(not a), b)), " |")
```

### **Converse : [(¬B) →A]**

```python
print("|  A  |  B  |  ¬B  |  S  |", )

for i in range(0, 4):
    a = i // 2
    b = i % 2
    print("| ", a, " | ", b, " |  ", int(not b), " | ", (conditional(int(not b), a)), " |")
```

## **Contrapositive : [B → (¬A)]**

```python
print("|  A  |  B  |  ¬A  |  S  |", )

for i in range(0, 4):
    a = i // 2
    b = i % 2
    print("| ", a, " | ", b, " |  ", int(not a), " | ", (conditional(b, int(not a))), " |")
```
